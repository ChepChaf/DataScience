### Do students that majored in subjects that were majority female make more money?

```python
more_women = recent_grads[recent_grads['Women'] > recent_grads['Men']]
more_men = recent_grads[recent_grads['Women'] < recent_grads['Men']]
```

```python
fig, ax = plt.subplots(figsize=(7, 10))
#ax1 = fig.add_subplot(2, 1, 1)
#ax2 = fig.add_subplot(2, 1, 2)
labels = ['Women', 'Men']
ax.boxplot([more_women['Median'].values, more_men['Median'].values])
ax.set_xticklabels(labels, rotation=90)
ax.set_ylim(0, 120000)
ax.set_xlim(left=0)
ax.set_title('Median for majors by more women and more men')
ax.set_ylabel('Median')
    
plt.show()
```

#### Seeing the boxplot it is visible that majors with more men have a higher median.

### What percent of majors are predominantly male? Predominantly female?

```python
import numpy as np
```

```python
women_percent = (len(more_women) / len(recent_grads)) * 100
men_percent = (len(more_men) / len(recent_grads)) * 100

x = np.arange(2) 
fig, ax = plt.subplots()
ax.bar(x, [women_percent, men_percent], width=0.5, 
       align='center', tick_label=['Women', 'Men'])
ax.set_ylabel('Percentage')
ax.set_title('Percentage of predominance.')
```

#### The relation seems to be 55-45 with more majors with female predominance
