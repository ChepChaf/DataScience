```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
data.head()
```

```python
data.describe()
```

## Let's plot the percentage of each gender by year 

```python
fig, ax = plt.subplots()
ax.plot(data['Year'], data['Physical Sciences'], c='blue', label='Women')
ax.plot(data['Year'], 100 - data['Physical Sciences'], c='green', label='Men')
ax.tick_params(top='off', bottom='off', left='off', right='off')
for spine in ax.spines.values():
    spine.set_visible(False)
ax.legend(loc='upper right')
ax.set_title('Physical Sciences Gender Percentage')
```
